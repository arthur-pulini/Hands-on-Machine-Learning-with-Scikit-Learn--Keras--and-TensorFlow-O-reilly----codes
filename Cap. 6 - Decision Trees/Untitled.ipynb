**Decision Trees**

**Treinando e visualizando**

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

iris = load_iris()
X = iris.data[:, 2:]
y = iris.target

tree_clf = DecisionTreeClassifier(max_depth=2)
tree_clf.fit(X, y)
```

```python
from sklearn.tree import export_graphviz
import graphviz

def graph_plot(model, feature_names, class_names=None):
    dotData = export_graphviz(model,
                              out_file=None,
                              feature_names=feature_names,
                              class_names=class_names,
                              rounded=True,
                              filled=True
        )
    graph = graphviz.Source(dotData)
    graph.view()
```

```python
graph_plot(tree_clf, iris.feature_names[2:], iris.target_names)
```

**Estimando a probabilidade das classes**

```python
tree_clf.predict_proba([[5, 1.5]]) #comprimento e largura
```

**0%** para classe 0, setosa.

**90,74%** para classe 1, versicolor.

**9,25%** para classe 2, virginica.

```python
tree_clf.predict([[5, 1.5]])
```

**Usando Árvore de Decisão para regressão**

```python
import numpy as np

np.random.seed(42)
m = 200
X = np.random.rand(m, 1)
y = 4 * (X - 0.5) ** 2
y = y + np.random.randn(m, 1) / 10
```

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor(max_depth=2)
tree_reg.fit(X, y)
```

```python
graph_plot(tree_reg, ["x1"])
```

```python
tree_reg.predict([[0.6]])
```

```python
tree_reg2 = DecisionTreeRegressor(max_depth=3)
tree_reg2.fit(X, y)
```

```python
graph_plot(tree_reg2, ["x1"])
```

```python
tree_reg2.predict([[0.6]])
```

```python
tree_reg3 = DecisionTreeRegressor(min_samples_leaf=10)
tree_reg3.fit(X, y)
```

```python
tree_reg3.predict([[0.6]])
```
